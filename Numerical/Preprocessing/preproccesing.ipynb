```python
#Filling missing data with the fillna way
df2['age'].fillna(df2['age'].mean(), inplace=True)
df2['sex'].fillna(df2['sex'].mode()[0], inplace=True)
df2['region'].fillna(df2['region'].mode()[0], inplace=True)
df2['charges'].fillna(df2['charges'].mean(), inplace=True)
print(df2.isnull().sum())
df2
```

```python
#Drop Duplicates data
df_cleaned = df2.drop_duplicates()
df2
```

```python
#check if the all columns types is correct and dedicte it
df2['bmi'] = pd.to_numeric(df2['bmi'], errors='coerce')

df2
```

```python
#fill the missing data of Bmi
df2['bmi'].fillna(df2['bmi'].mean(),inplace=True)
df2
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
#Label Encoding for sex and smoker columns
df2["sex"] = le.fit_transform(df["sex"])
df2["smoker"] = le.fit_transform(df["smoker"])

#Encoding with Dummies way for the region column
df2 = pd.get_dummies(df2,columns=['region'],drop_first=False)
```

```python
df2["region_northeast"] = df2["region_northeast"].astype(int)
df2["region_northwest"] = df2["region_northwest"].astype(int)
df2["region_southeast"] = df2["region_southeast"].astype(int)
df2["region_southwest"] = df2["region_southwest"].astype(int)
```
