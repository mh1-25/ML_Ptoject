```python
import joblib
from sklearn.neighbors import KNeighborsRegressor

# Load data
X_train = joblib.load("../Preprocessing/X_train.pkl")
X_test = joblib.load("../Preprocessing/X_test.pkl")
y_train = joblib.load("../Preprocessing/y_train.pkl")
y_test = joblib.load("../Preprocessing/y_test.pkl")
```

# <font color="red">KNN Model</font>

```python
from sklearn.neighbors import KNeighborsRegressor # for Continous values

knn = KNeighborsRegressor(n_neighbors=5)
knn.fit(X_train, y_train)

Y_pred = knn.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error

# to determine the accuracy of the model
print("MAE:", mean_absolute_error(y_test, Y_pred))  # This means the average error in the forecast is about $115.
```

```python
from sklearn.metrics import r2_score
print("R2:", r2_score(y_test, Y_pred))
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(8,6))
plt.scatter(y_test, Y_pred)

m, b = np.polyfit(y_test, Y_pred, 1)
plt.plot(y_test, m*y_test + b, color='red')

plt.xlabel("Actual Charges")
plt.ylabel("Predicted Charges")
plt.title("Actual vs Predicted (KNN)")
plt.show()
```
